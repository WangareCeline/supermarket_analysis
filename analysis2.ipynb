```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
# if load_dotenv() returns true it has found the .env file
load_dotenv()
con = os.getenv("LOCALDB_URL")
```

```python
engine =create_engine(con)
```

```python
sql_query = """
    select * from sales.purchase_data pd; 
"""
```

```python
purchase_data = pd.read_sql(sql_query ,con= engine)
```

```python
purchase_data.head()
```

```python
sql_query = """
    select * from sales.transaction_data pd; 
"""
transaction_data = pd.read_sql(sql_query ,con= engine)
transaction_data.head()
```

```python
#checking for null values
transaction_data.isna().sum()
```

```python
# checking number of (rows, columns)
transaction_data.shape
```

```python
#checking datatypes and not-null counts
transaction_data.info()
```

```python
sql_query = """
-- joining the two tables to be one table using a left join
select * 
from sales.purchase_data pd 
left join sales.transaction_data td 
on pd."LYLTY_CARD_NBR" =td."LYLTY_CARD_NBR"; 
"""
combined_data = pd.read_sql(sql_query ,con= engine)
combined_data.head()
```

```python
# Drop duplicates
combined_data.drop_duplicates(subset=['LYLTY_CARD_NBR'])
```

```python
# Drop duplicates
combined_data.drop_duplicates()
```

```python
# MERGING TABLES 
# unlike joining, merging will not have duplicate unique columns
merged_data = transaction_data.merge(purchase_data, on='LYLTY_CARD_NBR', how='left')
```

```python
merged_data.head()
```

### Objectives
1. Top ten most purchased brands
2. Total sales by customer life stage
3. Average spend by customer category

#### TOP TEN BRANDS

```python
top_brands = merged_data.groupby('PROD_NAME')['PROD_QTY'].sum().sort_values(ascending=False).head(10)
top_brands
```

### SALES BY LIFESTAGE

```python
sales_by_life_stage = merged_data.groupby('LIFESTAGE')['TOT_SALES'].sum().sort_values(ascending=False)
sales_by_life_stage
```

### AVERAGE SPEND BY CATEGORY

```python
avg_spend_by_cust_category = merged_data.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].mean()
avg_spend_by_cust_category
```

### MONTHLY SALES TREND

```python
# GETTING A NEW COLUMN FROM EXISTING DATA 
merged_data['YEAR']= merged_data['DATE'].dt.year
merged_data['MONTH'] = merged_data['DATE'].dt.month
merged_data['MONTH_NAME'] = merged_data['DATE'].dt.month_name()
merged_data.head()
```

```python
monthly_sales = merged_data.groupby(['YEAR', 'MONTH'])['TOT_SALES'].sum().reset_index().sort_values(['YEAR', 'MONTH'])
monthly_sales
```

### STORE PERFORMANCE
TOTAL SALES BY STORE

```python
store_sales = merged_data.groupby('STORE_NBR')['TOT_SALES'].sum().sort_values(ascending=False).head(10)
store_sales
```

### PRODUCT PERFORMANCE

```python
product_sales = merged_data.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False).head(10)
product_sales
```

### CUSTOMER BEHAVIOR
Which lifestage buys the largest quantity

```python
quantity_by_lifastage = merged_data.groupby('LIFESTAGE')['PROD_QTY'].sum().sort_values(ascending=False)
quantity_by_lifastage
```

### Which premium customer spends the most overall?

```python
premium_customer = merged_data.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].sum().sort_values(ascending=False)
premium_customer
```

### Which customer lifestage accounts for the highest total sales among your "Premium" customer base. 

```python
top_premium_spender = merged_data.query("PREMIUM_CUSTOMER =='Premium'").groupby('LIFESTAGE')['TOT_SALES'].sum().sort_values(ascending=False)
top_premium_spender
```
