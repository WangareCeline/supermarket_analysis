```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
purchase_df= pd.read_csv("purchase_b.csv")
transaction_df= pd.read_excel("transaction.xlsx")
```

```python
from sqlalchemy import create_engine

#connecting to db hosted on machine
# Replace with your actual credentials and database name
db_user = 'postgres'
db_password = 'root'
db_host = 'localhost'
db_port = '5432' # Default PostgreSQL port
db_name = 'supermarket'

engine = create_engine(f'postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')
```

```python
#pushing files to the db on machine
transaction_df.to_sql("transaction_data", engine, if_exists='replace', index=False, schema='sales')
```

```python
#pushing files to the db on machine
purchase_df.to_sql("purchase_data", engine, if_exists='append', index=False, schema='sales')
```

```python
# connecting to localdb using .env file
from dotenv import load_dotenv
import os
load_dotenv()
localdb_url = os.getenv("LOCALDB_URL")

```

```python
engine = create_engine(localdb_url)
```

```python
#pushing files to the localdb
transaction_df.to_sql("transaction_data", engine, if_exists='append', index=False, schema='chipsales')
```

```python
#pushing files to the localdb
purchase_df.to_sql("purchase_data", engine, if_exists='append', index=False, schema='chipsales')
```
