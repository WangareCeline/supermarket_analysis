```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
# if load_dotenv() returns true it has found the .env file
load_dotenv()
localdb_url = os.getenv("LOCALDB_URL")
```

```python
sql_query = """
    -- joining the two tables to be one table using a left join
    select * 
    from chipsales.purchase_data pd 
    left join chipsales.transaction_data td 
    on pd."LYLTY_CARD_NBR" =td."LYLTY_CARD_NBR" ;
"""
```

```python
data_df = pd.read_sql(sql_query ,con= localdb_url)
```
