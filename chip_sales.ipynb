```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
db_url = os.getenv("DATABASE_URL")
```

```python
purchase_df =pd.read_csv('purchase_b.csv')
```

```python
# run pip install openpyxl in the terminal ndo pandas iweze kusoma excel file
transaction_df = pd.read_excel('transaction.xlsx')
```

```python
purchase_df.head
```

```python
transaction_df.tail
```

```python
engine = create_engine(db_url)
```

```python
transaction_df.to_sql('transaction_data', engine, if_exists='replace', index=False, schema='customer_info')
```

```python
purchase_df.to_sql('purchase_data', engine, if_exists='append', index=False,schema='customer_info')
```
